# Exercício 6: Análise de Dados dos Pinguins: Acesse a base de dados "Pinguins" disponível em PalmerPenguins GitHub. Identifique as 'features' e o 'target' na base de dados.

```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier




penguins_df = seaborn.load_dataset("penguins")  
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
X = penguins_df[features] 
y = (penguins_df['species'] == 'Adelie').astype(int)

print("\n Essas são as features:")
print(X.head())

print("\n esse é o target (species) para teremos Adelie=1, Outras=0:")
print(y.head())
```

# Exercício 7: Preparando os Dados:Com a base de dados "Pinguins", escreva um código em Python para separar os dados em conjuntos de treino e validação. Observe que iremos criar um classificador que diferencia a espécie "Adelie" das outras duas.

```python
test_size = 0.3
random_seed = 42


mask = X.notna().all(axis=1) & y.notna()
X = X[mask]
y = y[mask]


np.random.seed(random_seed)
# SPLIT DOS DADOS
X_train, X_test, y_train, y_test = train_test_split(
    X.to_numpy(), y.to_numpy(), test_size=test_size, stratify=y, random_state=random_seed
)



print(f"\nTamanho do conjunto de treino: {len(X_train)} observações")
print(f"Tamanho do conjunto de validação: {len(X_test)} observações")
 
```

# Exercício 8: Construção do Modelo Inicial: Utilizando a biblioteca Scikit-Learn, escreva um código em Python para construir um modelo de Machine Learning usando o algoritmo de K-Nearest Neighbors com a base de dados "Pinguins".

```python
# normalizando: 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_scaled = scaler.transform(X)
X_test_scaled = scaler.transform(X_test)


distance_metric = 'minkowski'
weights = 'uniform'  

# TREINAR O MODELO
knn = KNeighborsClassifier(n_neighbors=5, metric=distance_metric, n_jobs=-1, weights=weights)
knn.fit(X_train_scaled, y_train)
```

# Exercício 9: Avaliação do Modelo:Após treinar o modelo de K-Nearest Neighbors com a base de dados "Pinguins", escreva um código para avaliar a acurácia do seu modelo nos dados de validação.

```python


train_accuracy = knn.score(X_train_scaled, y_train)
test_accuracy = knn.score(X_test_scaled, y_test)

print(f"\nAcurácia no treino: {train_accuracy*100:.2f}%")
print(f"Acurácia no teste: {test_accuracy*100:.2f}%")
```
